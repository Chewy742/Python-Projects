````python
def my_transliterate(w, source, target, diagraph = False):
    done = False
    result = []
    i = 0
    
    if diagraph != True and diagraph != False:
        raise ValueError("Please give a boolean for diagraph. No input specifies False.")

    
    if type(w) is not str: 
        raise ValueError("Please give a string to transliterate. ")
        
    if len(target) != len(source): 
        raise ValueError("Your source and target lists were not of the same length.")
        
    ##Iterate through the string to be transliterated
    while i < len(w):
        
        toAdd = None
        
        ##Check if the character is associated with a 2 letter transliteration
        if diagraph: 
            currentletter = w[i:i+2]
            try: 
                currentpos = source.index(currentletter) ##index of the character of character in w
                toAdd = target[currentpos]
                i += 2
            except ValueError as e: ##if this character wasn't a 1 to 2 character swap
                toAdd = None
        
        ##otherwise just transliterate the characters 1 to 1
        if toAdd is None:
            currentletter = w[i]
            try: 
                currentpos = source.index(currentletter)
                toAdd = target[currentpos]
                i += 1
            except ValueError as e: ##Error Handling
                raise ValueError("Not a valid chat string at position ", i, "symbol= ", w[i])
            except Error as e:
                print("Sorry, an " , e, " error occurred. Please try again with acceptable inputs")
                
                
            
        result.append(toAdd)
        
    result = "".join(result)
    return result
````

## Question 2: Parity Bit/Error Correcting 

__Overview__

- "A parity bit, or check bit, is a bit added to a string of binary code to ensure that the total number of 1-bits in the string is even or odd. Parity bits are used as the simplest form of error detecting code." (https://en.wikipedia.org/wiki/Parity_bit)


- A byte is 8-bits. Various forms of physical interference or errors can corrupt a bit in transmission, in memory, on a disk, etc. Assume that a bit can be in one of 3 states: 0, 1, U. U indicates that 0/1 state of the bit is uncertain.


- Implement two Python classes
    - ```CorrectingByte``` implements the core logic for a parity correcting byte.
        - The constructor raises an exception if the data fails a parity check or has more than one 'U' value. If that data has a single 'U', the constructor corrects the data or parity bit.
        - The method ```set_byte(b)``` sets the object's value. The method:
            - Sets the value if the byte passes parity check or is has a single 'U'. 
            - If the data has a single 'U', the method corrects the byte or parity bit.
            - Raise the exception is the byte has more than one 'U' or fails parity check.
        - ```get_byte()``` returns the correct value of a byte or raises an exception if the byte has become corrupted while stored.
        - ```flip_bit(i, v, p)``` allows a program to simulate a random change in a bit's value. ```p``` is an optional value. If ```p``` is not ```None``` the method sets the parity bit to ```v```. If ```p``` is ```None,``` the method sets the i-th bit in the byte to ```v.```
    - ```AbstractBit``` has a constructor, and methods ```set_bit(v)``` and ```get_bit().``` The class ensures that a bit's value is one of 1, 0, U. The class may require additional methods.



__Implementation__

````python
class AbstractBit(object):
    
    values = ['0', '1', 'U'] ##Three possible values of bit
    
    def __init__(self, v): 
         self.set_bit(v)
            
    def get_bit(self):
        return self.value
    
    def set_bit(self, v):
        if v not in AbstractBit.values: ##Error handline
            raise ValueError("Your value is not a valid value for a bit.")
        else: 
            self.value = v
            
    def __eq__(self, other):
        return self.get_bit() == other.get_bit()
    
    def __str__(self):
        
        return str(self.get_bit())
    
    
````

````python
class CorrectingByte(object):
    
    bitnumber = 8
    
    def __init__(self, b):
        
        if type(b[0]) is not str or type(b[1]) is not str:
            raise ValueError("Please give a string specifying a byte with 8 bits and a separate string specifying the parity bit")
            
        if(len(b[0]) != CorrectingByte.bitnumber): ##Has to have 8 bits
            raise ValueError("Incorrect number of bits")
        
        
        ##2 variables- A list of Abstract Bits (b), and 
        self.b = []
        
        for p in b[0]: 
            self.b.append(AbstractBit(p))
            
        self.p = AbstractBit(b[1])
        self.correct_byte()
            
    def correct_byte(self):
        
        ##Case where there are too many U bits
        count = 0
        for x in self.b:
            if (x.get_bit() == 'U'):
                count += 1
                
        if (count > 1 or (self.p.get_bit() == 'U' and count > 0)):
            raise ValueError("There are too many U bits in this byte")
            
        ##Next- where the p is U
        if (self.p.get_bit() == 'U'):
            newp = self.compute_parity()
            self.p = AbstractBit(str(newp))
            
        ##There is a U in the byte- this should always fix the byte with a non-U parity- woohoo i found a bug with his code!!
        
        if (count == 1):
            pos = self.b.index(AbstractBit("U"))
            self.b[pos] = AbstractBit("0") ##Try 0 first
            p = self.compute_parity()
            bit = self.p.get_bit()
            if (p != int(bit)):
                self.b[pos] = AbstractBit("1") ##otherwise its 1
                return (self.b, self.p)
            return (self.b, self.p)
            
        ##There are multiple corrections that could be made
        p = self.compute_parity()
        bit = self.p.get_bit()
        if p != int(bit):
            raise ValueError("Byte and parity corrupted due to multiple errors")
        
        return (self.b, self.p)
    
    def compute_parity(self): ##Is it even or odd
        
        v = 0
        for x in self.b:
            v += int(x.get_bit())
        
        v = v % 2
        return v 
            
    def set_byte(self, b):
        
        self.b = []
        
        for p in b[0]:
            self.b.append(AbstractBit(p))
            
        self.p = AbstractBit(b[1])
        self.correct_byte()
            
        
    def get_byte(self):
        self.correct_byte()
        return str(self)
    
    def flip_bit(self, i, v, p = None):
        if i is not int: 
            raise ValueError("Please specify an acceptable integer position for the bit you want to flip. Run again. :")
        if p is not None:
            self.p = AbstractBit(v)
        else:
            self.b[i] = AbstractBit(v)
            
    def __str__(self):
        
        s = "("
        b = len(self.b)
        for x in self.b:
            s += str(x)
        s += "," + str(self.p) + ")"
        
        return s 
        
````

## Question 3: Computing Football Table

__Overview__

- "The Premier League is the top level of the English football league system. ... ... There are 20 clubs in the Premier League. During the course of a season (from August to May) each club plays the others twice (a double round-robin system), once at their home stadium and once at that of their opponents', for a total of 38 games. Teams receive three points for a win and one point for a draw. No points are awarded for a loss. Teams are ranked by total points, then goal difference, and then goals scored." (https://en.wikipedia.org/wiki/Premier_League)


- The file epl2016.csv contains game-by-game results from the 2016-2017 season. The file contains many columns. Only the following contribute to the computation of final standings:
    - HomeTeam
    - AwayTeam
    - FTHG: The number of goals the home team scored.
    - FTAG: The number of goals the away team scored.
    - FTR: Is 'H', if the home team won, 'A' if the away team won and 'D' if the match was a draw. 

<u>Tasks</u>

You will implement several functions. Your functions will use Pandas and matplotlib.pyplot.

- ```load_epl_games(file_name)``` loads the file and creates a Pandas data frame containing the contents.


- ```get_points(team, table)``` takes a team name and a Pandas data frame. The method returns the number of points that the team earned.


- ```get_finish(team, table)``` returns the finishing position of a team.


- ```compute_table(df)``` uses the finishing position rules to return the final table.
    - The table is sorted by the finishing position rules.
    - The columns are ```Team,``` ```Points,``` ```Goals``` and ```GoalDifference.```
    
    
- ```display_hbar(tables)``` displays a horizontal bar chart.
    - The bar chart is sorted from final position from 1st place to last place.
    - The bar is the number of points.
    

````python
import pandas as pd
import matplotlib.pyplot as plt 

def load_epl_games(file_name): ##using pandas to read a file
    try: 
        footballdf = pd.read_csv(file_name)
        return footballdf
    except FileNotFoundError as e: 
        print("Your file was not found: ", e)

def compute_table(df):
    copy = df[["Date", "HomeTeam", "AwayTeam", "FTR", "FTHG", "FTAG"]].copy()
    result = pd.DataFrame()
    
    
    
    ##table columns
    result['Team'] = get_team_names(df)
    result['Points'] = 0 
    result['GoalDifference'] = 0
    result['Goals'] = 0
    result = result.set_index(["Team"])
    
    
    ##Computes points for each result
    for r in copy.itertuples():
        if r.FTR == "H":
            result.loc[r.HomeTeam, "Points"] += 3 #win
        elif r.FTR =="A":
            result.loc[r.AwayTeam, "Points"] += 3
        else:
            result.loc[r.HomeTeam, "Points"] += 1 #draw
            result.loc[r.AwayTeam, "Points"] += 1 
            
        result.loc[r.HomeTeam, "Goals"] += r.FTHG
        result.loc[r.HomeTeam, "GoalDifference"] += r.FTHG - r.FTAG ##Goal difference is goals for - goals for away team
        result.loc[r.AwayTeam, "Goals"] += r.FTAG
        result.loc[r.AwayTeam, "Goals"] += r.FTAG - r.FTHG ##opposite for away team
        
    result = result.sort_values(by = ['Points', "GoalDifference", "Goals"], ascending = False) ##sorting
    finish = 1
    result["Finish"] = 0
    for r in result.index: ##indexing the table from 1 - 20 
        result.loc[r, "Finish"] = finish 
        finish += 1
        
    return result


def get_team_names(table): 
    
    return table["HomeTeam"].unique()

def get_points(team, table):
    try:
        return table.loc[team, "Points"]
    except KeyError as e:
        print("Please give a valid team name for the 16-17 season. Returning none. ")
    except Error as f:
        print("An error occurred, make sure your inputs are valid! ", f)

def get_finish(team, table):
    try:
        return table.loc[team, 'Finish']
    except KeyError as e:
        print("Please give a valid team name for the 16-17 season. Returning none. ")
    except Error as f:
        print("An error occured, make sure your inputs are valid! ", f)
        
def get_goals_scored(team, table):

    try:
        return table.loc[team, 'Goals']
    except KeyError as e:
        print("Please give a valid team name for the 16-17 season. Returning none. ")
    except Error as f:
        print("An error occured, make sure your inputs are valid! ", f)
        


def display_hbar(table): ##displaying the horizontal bar chart in a cool way, courtesy of Prof Ferguson
    fig, ax = plt.subplots(figsize= (10,10))
    t = table.index
    t = list(t)
    t.reverse()
    p = list(table["Points"])
    p.reverse()
    ax.barh(t, p, align = 'center', alpha = .5)
    
    plt.ylabel("Team")
    plt.xlabel("Points")
    ax.tick_params(labeltop = True, bottom=True)
    ax.set_xticks(range(0, 100, 10))
    for d2 in range(0, 100, 5):
        plt.axvline(x = d2, color = "k", linestyle = '--')
        
    plt.show()
    


games = load_epl_games('epl2016.csv')
final_table = compute_table(games)
````
